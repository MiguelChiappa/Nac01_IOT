```python
import cv2 as cv
import numpy as np
import math
from matplotlib import pyplot as plt

img = cv.imread("circulo.png")

grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

circles = cv.HoughCircles(grey, cv.HOUGH_GRADIENT, dp=1.1, minDist=500, param1=200, param2=40, minRadius=50, maxRadius=400)

circles = np.uint16(np.around(circles))

x_min = circles[0][0][0]
x_max = circles[0][1][0]

y_max = circles[0][0][1]
y_min = circles[0][1][1]

cimg = img.copy()

#RÚBRICA 1:

z=0
for i in circles[0, 0:2]:
    # Segmenta apenas os círculos maiores e calcula a área dos mesmos:
    cv.circle(cimg, (i[0], i[1]), i[2], (255, 0, 200), 10)
    z=z+1
    print("Area " + str(z) + ": " + str(math.sqrt(i[2])*3.14))

    # Mostra o centro de massa circulo:
    cv.circle(cimg, (i[0], i[1]), 2, (255, 0, 200), 15)
    
#rÚBRICA 2:    

    # Traça uma linha entre os centros de massa:     
    plt.plot([x_min, x_max], [y_max, y_min])

# Calcula o ângulo de inclinação da reta em relação ao plano horizontal:
rad_to_deg=57.29578
deltaX = x_min-x_max 
deltaY = y_max-y_min
radians = math.atan2(deltaY, deltaX)
degrees = radians * rad_to_deg
print("Ângulo de dentro: ", degrees)
print("Ângulo suplementar: ", 180 - degrees) 


plt.imshow(cimg)

plt.show()
```
